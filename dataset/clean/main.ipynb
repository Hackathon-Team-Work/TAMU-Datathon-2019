```
# "common" ingredients in tacos/burritos?
veg_blacklist = ['beef', 'chicken', 'pork', 'carne', 'carni', 'ham', 'fish']
# vegans are vegetarian + no dairy
vegan_blacklist = veg_blacklist + ['egg', 'milk', 'yogurt', 'yoghurt', 'sour cream', 'cream']

def is_vegan(row):
    desc_col = row['menus.description']
    # if the value is NOT NaN
    if desc_col != np.nan and isinstance(desc_col, str):
        # LOWERCASE it
        row['menus.description'] = row['menus.description'].lower()
        for ingred in vegan_blacklist:
            if ingred in row['menus.description']:
                # found a bad ingred., fail
                return False
        # didn't find anything bad, must be good
        return True
    else:
        # assume the worst...
        return False
    
def is_veg(row):
    # comments, code is same as above!
    desc_col = row['menus.description']
    if desc_col != np.nan and isinstance(desc_col, str):
        row['menus.description'] = row['menus.description'].lower()
        for ingred in veg_blacklist: # except veg. blacklist!
            if ingred in row['menus.description']:
                return False
        return True
    else:
        return False
    
sample_df['is_vegan'] = sample_df.apply (is_vegan, axis=1)
sample_df['is_veg'] = sample_df.apply (is_veg, axis=1)
```
